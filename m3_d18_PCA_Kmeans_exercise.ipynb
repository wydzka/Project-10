- CASHADVANCEFREQUENCY: How frequently the cash in advance being paid
- CASHADVANCETRX: Number of Transactions made with "Cash in Advanced"
- PURCHASESTRX: Numbe of purchase transactions made
- CREDITLIMIT: Limit of Credit Card for user
- PAYMENTS: Amount of Payment done by user
- MINIMUM_PAYMENTS: Minimum amount of payments made by user
- PRCFULLPAYMENT: Percent of full payment paid by user
- TENURE: Tenure of credit card service for user

## **Data Preprocessing**

### **Drop unused columns**

Karena Customer ID adalah ID unik untuk tiap customer, maka fitur ini tidak memiliki peranan penting untuk penentuan cluster nantinya.

```python
df = pd.read_csv('creditcard.csv')
```

```python
df.drop(['CUST_ID'], axis=1, inplace=True)
```

### **Drop missing value pada CREDIT_LIMIT**



```python
df.dropna(subset=['CREDIT_LIMIT'], inplace=True)
```

### **Imputasi missing value pada MINIMUM_PAYMENTS**

Karena distribusi pada MINIMUM_PAYMENTS tidak mengikuti distribusi normal (terdapat skewness), maka mengimputasi missing value dengan menggunakan nilai median akan memberikan estimasi nilai yang lebih baik.

```python
# df['CREDIT_LIMIT'].fillna(df['CREDIT_LIMIT'].median(), inplace=True)
```

```python
df['MINIMUM_PAYMENTS'].fillna(df['MINIMUM_PAYMENTS'].median(), inplace=True)
```

### **Scaling data**

- scaling dibutuhkan pada PCA karena menghitung jarak

### **Hanya menggunakan 2 Principal Component**

```python
# 2 principal components
pca = PCA(n_components=2, random_state=0)

# fit dan transform
X_pca = pca.fit_transform(df_scaled)
```

```python
# Hanya 2 kolom yang akan kita gunakan untuk clustering nanti
X_pca
```

```python
# Variance denga 2 PC
df_var = pd.DataFrame({
    'PC': range(X_pca.shape[1]),
    'variance': pca.explained_variance_ratio_,
    'cumsum_variance': pca.explained_variance_ratio_.cumsum()
})

df_var
```

```python
list_imp = []

for i in range(pca.components_.shape[0]): 

    pc_i_importance = pd.DataFrame({
        'feature': X.columns,
        'coef': np.abs(pca.components_[i])
    }).sort_values('coef',ascending=False)

    imp = pc_i_importance.head(3)['feature'].to_list() # 3 feature terpenting pada suatu PC
    list_imp.append(imp)

# -------------------------------------------
df_var['feature_importance'] = list_imp
df_var 
```

```python
[i  for i in df_var['feature_importance']]
```

### **Interpretasi**

- Dari proses PCA di atas, kita hanya akan menggunakan 2 Principal Components untuk nantinya digunakan pada Clustering. 
- Variance yang dapat dijelaskan oleh 2 Principal Components ini sebesar 0.47

## **K-Means clustering**


```python
list_k = range(2,10)

list_inertia = []
list_sil = []

for k in list_k:

    kmeans = KMeans(n_clusters= k, random_state=0)
    kmeans.fit(X_pca)

    # inertia untuk elbow mwthod
    inertia = kmeans.inertia_
    list_inertia.append(inertia)

    # silhouette score
    sil = silhouette_score(X_pca, kmeans.labels_)
    list_sil.append(sil)
```

```python
df_iner = pd.DataFrame({
    'k': list_k,
    'inertia': list_inertia,
    'sil_score': list_sil
})

df_iner 
```

### **Elbow method**

```python
# Within Cluster Sum of Square (inertia) dari tiap model dengan jumlah cluster berbeda
sns.scatterplot(data=df_iner, x='k', y='inertia')
sns.lineplot(data=df_iner, x='k', y='inertia')

plt.xlabel('Jumlah cluster')
plt.ylabel('Inertia')
plt.title('Elbow method');
```

**Interpretasi**

Dari Elbow Method, belum terlihat letak "elbow"nya untuk menentukan jumlah cluster terbaik

### **Silhouette method**

```python
df_iner.sort_values('sil_score', ascending=False)
```

```python
# looping nilai silhotte score untuk tiap penggunaan jumlah cluster berbeda 
sns.scatterplot(data=df_iner, x='k', y='sil_score')
sns.lineplot(data=df_iner, x='k', y='sil_score')

plt.title('Silhoutte scores vs Number of clusters')
plt.xlabel('Number of clusters')
plt.ylabel('Silhoutte score');
```

**Interpretasi**

Berdasarkan silhouette score-nya, jumlah cluster terbaik untuk dataset ini adalah 2 clusters. 

### **Hasil Clustering dengan 2 cluster**

```python
# Kmean dengan jumlah cluster 2
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(X_pca)

# label yang diberikan pada tiap data point dengan jumlah cluster 2 (0 dan 1)
cluster = kmeans.labels_

# contoh
cluster[0:100]
```

```python
# hasil KMeans clustering dengan jumlah cluster 2

pca_df = pd.DataFrame(data=X_pca, columns=['PC0', 'PC1'])

pca_df['labels'] = cluster
pca_df.head()
```

```python
# Visualisasi Cluster-0 (biru) dan Cluster-1 (orange)
plt.figure(figsize=(10,10))
sns.scatterplot(x='PC0', y='PC1', hue='labels', data=pca_df, palette='bright');
```
